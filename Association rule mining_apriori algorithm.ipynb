```python
import time
start = time.time()
df2 = pd.read_excel("dataprep3.xlsx")
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

frequent_itemsets = apriori(df2, min_support=0.03, use_colnames=True)
#frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets


rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
rules['length'] = rules['antecedents'].apply(lambda x: len(x))
end = time.time()

print("Time Taken is:")
print(end-start)
```

```python
import re
b=pd.DataFrame()
d=[]
a=0
a1=0
b=rules[(rules['length']>=2)]
b.index=range(0,len(b))
for i in b['consequents']:
    if ('rs>166' in i or 'rs140-166' in i or 'rs<140' in i) and len(i)==1:      
        d.append(b.loc[a1])
    a1=a1+1


d1=pd.DataFrame()
d1=pd.DataFrame(d)

d1.sort_values(by='confidence', ascending=False)
```
